```python
# agg method
# passing dict
genres.agg(                                    # on which genre we want to apply which aggregate function
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
genres[['Runtime','IMDB_Rating']].agg(['min','max','mean','sum'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# # looping on groups
# df = pd.DataFrame(columns=movies.columns) # empty dataframe
# for group,data in genres:
#   df = df.concat(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

# df
```

```python
# looping on groups  by Gemini
# Reset Index: We reset the index of the row_to_append DataFrame using reset_index(drop=True).
# This ensures that the index starts from 0 and avoids potential conflicts with the existing index of df.
# Ignore Index: The ignore_index=True argument in pd.concat tells pandas to create a new index for the resulting DataFrame,
# preventing duplicate indices
df = pd.DataFrame(columns=movies.columns)
for group, data in genres:
  row_to_append = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
  # Reset index to avoid potential index mismatches
  row_to_append = row_to_append.reset_index(drop=True)
  df = pd.concat([df, row_to_append], ignore_index=True)  # ignore_index to avoid duplicate indices

df
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo) # foo is returning a group and this apply function will return complete dataframe after adding all the data frames
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
