```python
# import titanic
df = pd.read_csv('/content/titanic.csv')
df['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
# len
df['Name'][df['Name'].str.len() == 82].values[0]
# strip
"                   vishal                              ".strip( )
df['Name'].str.strip() # remove leading and trailing spaces
```

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True) # n specifies how much split i have to do , expand se ye dataframe banakar return karega
df.head()

df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.endswith('A')]
# isdigit/isalpha...
df[df['firstname'].str.isdigit()]
```

```python
# applying regex
# contains
# search john -> both case
df[df['firstname'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```

```python
# slicing
df['Name'].str[::-1]
```
